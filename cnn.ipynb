```python
import tensorflow as tf


IMG_WIDTH = 224
IMG_HEIGHT = 224
IMG_CHANNELS = 1

#Build the model
inputs = tf.keras.layers.Input((IMG_HEIGHT, IMG_WIDTH,IMG_CHANNELS))
s = tf.keras.layers.Lambda(lambda x: x / 255)(inputs) ###convert to float  (normalizing the pixel values)

#conv layer 1
c1 = tf.keras.layers.Conv2D(32, (4, 4), activation='relu', kernel_initializer='he_normal')(s)
c1 = tf.keras.layers.Dropout(0.2)(c1)
c1 = tf.keras.layers.ZeroPadding2D(padding=(3,3))(c1)
p1 = tf.keras.layers.MaxPooling2D((2, 2))(c1)
#conv layer 2
c2 = tf.keras.layers.Conv2D(64, (5, 5), activation='relu', kernel_initializer='he_normal')(p1)
c2 = tf.keras.layers.Dropout(0.2)(c2)
c2 = tf.keras.layers.ZeroPadding2D(padding=(2,2))(c2)
p2 = tf.keras.layers.MaxPooling2D((2, 2))(c2)
#conv layer 3
c3 = tf.keras.layers.Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_normal')(p2)
c3 = tf.keras.layers.Dropout(0.2)(c3)
c3 = tf.keras.layers.ZeroPadding2D(padding=(1,1))(c3)
p3 = tf.keras.layers.MaxPooling2D((2, 2))(c3)
#conv layer 4 
c4 = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal')(p3)
c4 = tf.keras.layers.Dropout(0.2)(c4)
c4 = tf.keras.layers.ZeroPadding2D(padding=(1,1))(c4)
p4 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(c4)
#conv layer 5
c5 = tf.keras.layers.Conv2D(512, (3, 3), activation='relu', kernel_initializer='he_normal')(p4)
c5 = tf.keras.layers.Dropout(0.2)(c5)
c5 = tf.keras.layers.ZeroPadding2D(padding=(1,1))(c5)
p5 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(c5)

# flatten
flatten=tf.keras.layers.Flatten()(p5)
############### 6 fully connected layers#################
f1=tf.keras.layers.Dense(1024,activation='relu')(p5)
f2=tf.keras.layers.Dense(1024,activation='relu')(f1)
f3=tf.keras.layers.Dense(32,activation='relu')(f2)
f4=tf.keras.layers.Dense(16,activation='relu')(f3)
f5=tf.keras.layers.Dense(1024,activation='relu')(f4)
outputs = tf.keras.layers.Conv2D(3, (1, 1), activation='softmax')(f5)
 
model = tf.keras.Model(inputs=[inputs], outputs=[outputs])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```
