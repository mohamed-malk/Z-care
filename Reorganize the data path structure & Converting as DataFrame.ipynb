# Reorganize the data path structure

ADNI DataSet/class/ADNI/subject_id/serial_description/date/sample_id/image_name.nii
ADNI DataSet - Handled/class/serial_description/image_name.nii

```python
path = '/content/drive/MyDrive/Graduation Project Repo/ADNI DataSet'

columns = ['name', 'old path', 'new path']

ad_paths = pd.DataFrame(columns=columns)
cn_paths = pd.DataFrame(columns=columns)
mci_paths = pd.DataFrame(columns=columns)
```

```python
def handle(class_name):
  """
  Reorganize the data path structure 
  from 
       Dataset ->
                 
  """
  names, old_paths, new_paths =  [], [], []
  subs = os.listdir(path + '/' + class_name + '/ADNI')
  for sub in subs:
      types = os.listdir(path + '/' + class_name + '/ADNI' + '/' + sub)
      for type_ in types:
          dates = os.listdir(path + '/' + class_name + '/ADNI' + '/' + sub + '/' + type_)
          for date in dates :
              samples = os.listdir(path + '/' + class_name + '/ADNI' + '/' + sub + '/' + type_ + '/' + date)
              for sample in samples :
                  imgs = os.listdir(path + '/' + class_name + '/ADNI' + '/' + sub + '/' + type_ + '/' + date + '/' + sample)
                  for img in imgs:
                      img_path = path + '/' + class_name + '/ADNI' + '/' + sub + '/' + type_ + '/' + date + '/' + sample + '/'
                      #___________________
                      names.append(img); old_paths.append(img_path)
                      new_paths.append('ADNI DataSet - Handled/' + class_name  + '/' + type_ + '/')
                      #__________________

                      try:
                          os.mkdir('/content/drive/MyDrive/Graduation Project Repo/ADNI DataSet - Handled/' + class_name  + '/' + type_)
                      except FileExistsError:
                          pass

                      shutil.copy(img_path + img, 'ADNI DataSet - Handled/' + class_name  + '/' + type_ + '/' + img)

  return names, old_paths, new_paths
```

```python
names, old_paths, new_paths = handle('AD_T2')
ad_paths['name'] = names
ad_paths['old path'] = old_paths
ad_paths['new path'] = new_paths
ad_paths.to_csv('ADNI DataSet - Handled/AD_new_paths.csv')

names, old_paths, new_paths = handle('CN_T2')
cn_paths['name'] = names
cn_paths['old path'] = old_paths
cn_paths['new path'] = new_paths
cn_paths.to_csv('ADNI DataSet - Handled/CN_new_paths.csv')

names, old_paths, new_paths = handle('MCI_T2')
mci_paths['name'] = names
mci_paths['old path'] = old_paths
mci_paths['new path'] = new_paths
mci_paths.to_csv('ADNI DataSet - Handled/MCI_new_paths.csv')
```
